### ALS
Các siêu tham số quan trọng trong ALS bao gồm:

- maxIter: số lần lặp tối đa để chạy (mặc định là 10)
- rank: số lượng yếu tố ẩn trong mô hình (mặc định là 10)
- regParam: tham số điều chuẩn trong ALS (mặc định là 1.0)
